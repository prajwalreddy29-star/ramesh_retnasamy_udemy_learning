### **Create External Locations for Formula1 Project**

**Steps:**
1. Check existing external locations to find your S3 base path and storage credential name
2. Create external locations for different folders (raw, processed, presentation)
3. Verify the external location was created successfully

**Note:** Storage credentials should already exist in your workspace.

````sql
-- Step 1: Check existing external locations
-- Note the storage_credential_name and url columns to get your base S3 path
SHOW EXTERNAL LOCATIONS;
````

### **External Location Setup - Reference Documentation**

#### **Note:**
No need to create extra external locations as already I have created parent bucket as external location which should not be created in real project.

#### **Syntax: Create External Location (For Reference)**
```sql
-- Get the base S3 path from SHOW EXTERNAL LOCATIONS above, then create new location

CREATE EXTERNAL LOCATION IF NOT EXISTS formula1_raw_data
URL 's3://your-bucket-name/formula1/raw/'
WITH (STORAGE CREDENTIAL <existing_credential_name>)
COMMENT 'External location for Formula1 raw data';

-- Examples for different folders:
CREATE EXTERNAL LOCATION IF NOT EXISTS formula1_processed_data
URL 's3://your-bucket-name/formula1/processed/'
WITH (STORAGE CREDENTIAL <existing_credential_name>);

CREATE EXTERNAL LOCATION IF NOT EXISTS formula1_presentation_data
URL 's3://your-bucket-name/formula1/presentation/'
WITH (STORAGE CREDENTIAL <existing_credential_name>);
```

#### **Verify External Location (For Reference)**
```sql
-- Verify the external location
DESCRIBE EXTERNAL LOCATION formula1_raw_data;

-- Grant permissions if needed
GRANT CREATE EXTERNAL TABLE ON EXTERNAL LOCATION formula1_raw_data TO `<user@example.com>`;
```

````sql
SHOW CATALOGS;
````

````sql
-- Syntax: Create a Unity Catalog
CREATE CATALOG IF NOT EXISTS formula1
MANAGED LOCATION 's3://databricks-practice-bucket-01/formula1/'
COMMENT 'Formula1 main catalog';
````

````sql
-- Syntax: Create schemas within the catalog
-- First, use the catalog
USE CATALOG formula1;

CREATE SCHEMA IF NOT EXISTS landing;

-- Create bronze schema (raw/landing data)
CREATE SCHEMA IF NOT EXISTS bronze
MANAGED LOCATION 's3://databricks-practice-bucket-01/formula1/bronze/'
COMMENT 'Bronze layer - raw data ingestion';

-- Create silver schema (cleaned/transformed data)
CREATE SCHEMA IF NOT EXISTS silver
MANAGED LOCATION 's3://databricks-practice-bucket-01/formula1/silver/'
COMMENT 'Silver layer - cleansed and conformed data';

-- Create gold schema (business-level aggregates)
CREATE SCHEMA IF NOT EXISTS gold
MANAGED LOCATION 's3://databricks-practice-bucket-01/formula1/gold/'
COMMENT 'Gold layer - business-level aggregates';

-- Describe a specific schema
DESCRIBE SCHEMA EXTENDED formula1.bronze;
````

````sql
-- Verify schemas
SHOW SCHEMAS IN formula1;
````

````sql
DESCRIBE SCHEMA EXTENDED formula1.landing;
````

````sql
-- Syntax: Create External Volume for landing folder
-- External volumes provide access to external storage for file operations

-- Create external volume in landing schema
CREATE EXTERNAL VOLUME IF NOT EXISTS formula1.landing.files
LOCATION 's3://databricks-practice-bucket-01/formula1/landing/'
COMMENT 'External volume for landing raw files';
````

````sql
-- Verify the volume
DESCRIBE VOLUME formula1.landing.files;

-- Example: List files in the volume
-- LIST '/Volumes/formula1/landing/files/';
````
